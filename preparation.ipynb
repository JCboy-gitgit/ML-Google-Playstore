```python
import numpy as np
import pandas as pd
```

```python
# Utils
def print_missing_values(df):
    print(f'Number of missing values:\n{df.isnull().sum()}')
```

```python
df = pd.read_csv('dataset/googleplaystore.csv')

print_missing_values(df)
```

```python
df.head()
```

```python
df.info()
```

### Cleaning Duplicated Application

```python
df['App'].value_counts()
```

```python
# Drop duplicates
df.drop_duplicates(subset='App', inplace=True)
df['App'].value_counts()
```

```python
# Reindexing the dataframe
df.reset_index(drop=True, inplace=True)
df.info()
```

### Cleaning Rating

```python
rating_count = df['Rating'].unique() 
print(rating_count)

outliers = df[df['Rating'] > 5]
outliers
```

```python
# Since the outlier is so extreme (entry are shifted to the left), we drop it.
df.drop(outliers.index, inplace=True)
df.info()
```

```python
# rating > 5 to NaN
df.loc[df['Rating'] > 5, 'Rating'] = np.nan
rating_count = df['Rating'].unique()
print(rating_count)
```

```python
# forward fill missing values in 'Rating' column
df['Rating'] = df['Rating'].ffill()
print_missing_values(df)
```

```python
df['Rating'].describe()
```

### Cleaning Price

```python
df['Price'] = df['Price'].str.replace('$', '')
price_count = df['Price'].unique()
print(price_count)
```

```python

df.loc[df['Price'] == 'Everyone', 'Price'] = '0'

price_count = df['Price'].unique()
print(price_count)
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

df.info()
```
