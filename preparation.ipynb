```python
import numpy as np
import pandas as pd
```

```python
# Utils
def print_missing_values(df):
    print(f'Number of missing values:\n{df.isnull().sum()}')
```

```python
df = pd.read_csv('dataset/googleplaystore.csv')

print_missing_values(df)
```

```python
df.head()
```

```python
df.info()
```

### Cleaning Duplicated Application

```python
df['App'].value_counts()
```

```python
# Drop duplicates
df.drop_duplicates(subset='App', inplace=True)
df['App'].value_counts()
```

```python
# Reindexing the dataframe
df.reset_index(drop=True, inplace=True)
df.info()
```

### Cleaning Rating

```python
rating_count = df['Rating'].unique() 
print(rating_count)

outliers = df[df['Rating'] > 5]
outliers
```

```python
# Since the outlier is so extreme (entry are shifted to the left), we drop it.
df.drop(outliers.index, inplace=True)
df.info()
```

```python
# rating > 5 to NaN
df.loc[df['Rating'] > 5, 'Rating'] = np.nan
rating_count = df['Rating'].unique()
print(rating_count)
```

```python
# forward fill missing values in 'Rating' column
df['Rating'] = df['Rating'].ffill()
print_missing_values(df)
```

```python
df['Rating'].describe()
```

### Cleaning Price

```python
df['Price'] = df['Price'].str.replace('$', '')
price_count = df['Price'].unique()
print(price_count)
```

```python

df.loc[df['Price'] == 'Everyone', 'Price'] = '0'

price_count = df['Price'].unique()
print(price_count)
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

df.info()
```

### Cleaning Category

```python
df['Category'].unique()
```

### Cleaning Reviews

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Reviews'].describe()
```

### Cleaning Install

```python
df['Installs'].unique()
```

```python
df['Installs'] = df['Installs'].str.replace('+', '').str.replace(',', '')
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df['Installs'].unique()
```

```python
df.info()
```

### Cleaning Size

```python
df['Size'].unique()
```

```python
wtf = df[df['Size'] == 'Varies with device']

# Count wtf entries
print(f'Number of "Varies with device" entries: {len(wtf)}')
```

```python
# Standardize 'Size' column
def convert_size(size):
    if size == 'Varies with device':
        return np.nan
    elif size.endswith('M'):
        return float(size[:-1]) * 1024  # Convert MB to KB
    elif size.endswith('k'):
        return float(size[:-1])  # Already in KB
    else:
        return np.nan  # Handle unexpected formats
    
df['Size'] = df['Size'].apply(convert_size)
df['Size'].describe()
```

```python
df.info()
```

### Cleaning Type

```python
df['Type'].value_counts(dropna=False)
```

```python
# Consistency check for 'Type' column
inconsistent_type = df[(df['Price'] > 0) & (df['Type'] != 'Paid') | (df['Price'] == 0) & (df['Type'] != 'Free')]
print(f'Number of inconsistent entries in "Type" column: {len(inconsistent_type)}')
```

```python
# Resolve NaN values in 'Type' column based on 'Price'
df.loc[df['Price'] > 0, 'Type'] = 'Paid'
df.loc[df['Price'] == 0, 'Type'] = 'Free'
```

```python
df['Type'].unique()
```

```python
inconsistent_type = df[(df['Price'] > 0) & (df['Type'] != 'Paid') | (df['Price'] == 0) & (df['Type'] != 'Free')]
print(f'Number of inconsistent entries in "Type" column: {len(inconsistent_type)}')
```
